To  name a few modelling techniques:

a. Linear regression: works for simple systems where the relationship between observable variables is linear, it can be extended to deal with more complex relationships.

*  this method fits in the more general __split-apply-combine__ pattern.

* __Apply__ a function to each group independently
* __Combine__ the results into a data structure

````python
data = {
    'product': ['Apple','Banana', 'Apple','Banana','Apple'],
    'Region': ['North','North','South','South','North'],
    'Sales': [100,200,150,300,250],
    'Quantity': [10, 3, 4, 5, 22]
}

df = pd.DataFrame(data)
df
````

# 3. Filtering Data

Another way to select a subset of data is by applying __Boolean indexing.__, which is commonly called as __filter__.

For instance, it we want to filter those vlaues less than or equal to 6.5. We use comparision technique to filter out the data.

````python
edu[edu['Value']> 6.5].head()
````

Let's break down the two pieces of code to understand what each one does and how they differ:

### 1. **`edu['Value'] > 6.5`**

This expression creates a **boolean mask**. Heres what happens:

- **`edu['Value']`**: This accesses the `"Value"` column from the `edu` DataFrame.
- **`> 6.5`**: This checks each entry in the `"Value"` column to see if it's greater than `6.5`.

The result is a Series of boolean values (`True` or `False`) corresponding to each row in the `"Value"` column:

- **`True`**: The condition (`Value > 6.5`) is met for that row.
- **`False`**: The condition is not met.

So, **`edu['Value'] > 6.5`** by itself simply tells you, for each row, whether the value in the `"Value"` column is greater than `6.5`.

### Example Output:
```python
0     False
1     False
2      True
3     False
4      True
...
Name: Value, dtype: bool
```

### 2. **`edu[edu['Value'] > 6.5]`**

This expression filters the `edu` DataFrame to include only the rows where the condition `edu['Value'] > 6.5` is `True`. Heres what happens:

- **`edu['Value'] > 6.5`**: As described above, this produces a boolean Series.
- **`edu[...]`**: This applies the boolean Series as a filter to the DataFrame `edu`.

The result is a new DataFrame containing only the rows where the `"Value"` column is greater than `6.5`.

### Example Output:
If your original DataFrame (`edu`) had rows where the `"Value"` column had values like 6.2, 7.1, 6.8, etc., this would return only those rows where the `"Value"` is greater than 6.5:

```python
    TIME       GEO  Value
2   2020  Germany    7.1
4   2020   France    6.8
...
```

### Summary:

- **`edu['Value'] > 6.5`**: Returns a boolean Series indicating whether each value in the `"Value"` column is greater than `6.5`.
- **`edu[edu['Value'] > 6.5]`**: Uses that boolean Series to filter the DataFrame `edu`, returning only the rows where the condition is `True`.

In other words:
- **First expression** (boolean mask): "Which rows meet the condition?"
- **Second expression** (DataFrame filtering): "Give me the rows that meet the condition."
